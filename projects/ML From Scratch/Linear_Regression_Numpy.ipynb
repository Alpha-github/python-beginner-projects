```python
from sklearn import datasets
import matplotlib.pyplot as plt
import numpy as np
```

```python
X,y = datasets.make_regression(n_samples=1000,n_features=1, bias =0, noise=5, random_state=20)
```

```python
plt.scatter(X,y)
```

```python
X.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
m = X_train.shape[0] 
X_train = np.append(X_train, np.ones((m,1)), axis=1)
y_train = y_train.reshape(m,1)
```

```python
a = np.linalg.pinv(np.matmul(X_train.T,X_train))
print(a.shape)
b = np.matmul(X_train.T,y_train)
w = np.matmul(a,b)
w.shape
```

```python
w
```

```python
X_test = np.append(X_test, np.ones((X_test.shape[0],1)), axis=1)
```

```python
y_pred = np.matmul(X_test,w)
y_pred[:5,:]
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test,y_pred)
```

```python
plt.plot(X_test[:,0],y_pred, color="green")
plt.scatter(X_test[:,0],y_test)
```
